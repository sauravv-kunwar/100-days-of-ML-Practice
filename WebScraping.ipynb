```python
import sqlite3

# Create/connect DB
conn = sqlite3.connect('demo.db')
c = conn.cursor()

# Create table
c.execute('''CREATE TABLE IF NOT EXISTS households
             (id INTEGER PRIMARY KEY, gapaname TEXT, owned INTEGER, total INTEGER)''')

# Insert data
c.execute("INSERT OR IGNORE INTO households VALUES (1, 'Demo Gaunpalika', 100, 150)")

# Query
c.execute("SELECT * FROM households")
print(c.fetchall())  # Output: [(1, 'Demo Gaunpalika', 100, 150)]

conn.commit()
conn.close()
```

```python
from lxml import objectify
path = "./Performance_MNR.xml"
with open(path) as f:
    parsed = objectify.parse(f)
root = parsed.getroot()
# root.INDICATOR returns a generator yielding each <INDICATOR> XML element. 
# For each record, we can populate a dictionary of tag names (like YTD_ACTUAL) to data values (excluding a few tags) by running the following code:
data = []
skip_fields = ["PARENT_SEQ", "INDICATOR_SEQ",
"DESIRED_CHANGE", "DECIMAL_PLACES"]
for elt in root.INDICATOR:
    el_data = {}
    for child in elt.getchildren():
        if child.tag in skip_fields:
            continue
        el_data[child.tag] = child.pyval
    data.append(el_data)
# Lastly, convert this list of dictionaries into a DataFrame:
perf = pd.DataFrame(data)
perf.head()

Reading XML data with Pandas
Example-2: Use read_xml to create pandas dataframe
perf2 = pd.read_xml(path)
perf2.head()

2.5 Handling Large Datasets
Key Concepts
. . .

Terminology: - Generator: Yields items one-by-one (memory efficient) - Iterator: Object to traverse collections lazily - Streaming: Process without loading all at once

30 / 33
```
