# üìå Linear Regression

## ‚úÖ What is Linear Regression?
**Linear Regression** is a **supervised machine learning algorithm** used to predict a **continuous numerical value**.  
It models the relationship between **independent variables (features)** and a **dependent variable (target)** by fitting a **straight line**.

It is mainly used when we want to answer:
> "How does `Y` change when `X` changes?"

Example applications:
- Predicting house price based on area
- Predicting salary based on years of experience
- Predicting placement package based on CGPA

---

## üéØ Objective of Linear Regression
The main objective is to find the **best-fit line** that minimizes the error between:
- **Actual values**
- **Predicted values**

This is usually done by minimizing **Mean Squared Error (MSE)**.

---

## üßÆ Linear Regression Equation

### 1Ô∏è‚É£ Simple Linear Regression
Used when there is **only one input feature**.

\[
y = mx + c
\]

Where:
- **y** = predicted output (target)
- **x** = input feature
- **m** = slope (how much y changes when x increases)
- **c** = intercept (value of y when x = 0)

---

## üìå Meaning of Slope and Intercept
### üîπ Slope (m)
Slope shows the **rate of change**:
- If `m` is positive ‚Üí `y` increases when `x` increases
- If `m` is negative ‚Üí `y` decreases when `x` increases

### üîπ Intercept (c)
Intercept is the predicted value of `y` when `x = 0`.

---

## üß† Types of Linear Regression

### ‚úÖ 1) Simple Linear Regression
- Uses **one feature**
- Example: `CGPA ‚Üí Placement`

Equation:
\[
y = mx + c
\]

---

### ‚úÖ 2) Multiple Linear Regression
- Uses **more than one feature**
- Example: `CGPA + Skills + Experience ‚Üí Placement`

Equation:
\[
y = m_1x_1 + m_2x_2 + ... + m_nx_n + c
\]

---

## ‚≠ê Advantages of Linear Regression
- Simple and easy to understand
- Works well for linear relationships
- Fast to train and predict
- Good baseline model for regression problems

---

## ‚ö†Ô∏è Limitations of Linear Regression
- Works best only when relationship is linear
- Sensitive to outliers
- Can perform poorly if data is complex or non-linear
- Multicollinearity (high correlation between features) affects performance in multiple regression

---

## üìå Common Evaluation Metrics
To check model performance:
- **Mean Absolute Error (MAE)**
- **Mean Squared Error (MSE)**
- **Root Mean Squared Error (RMSE)**
- **R¬≤ Score (Coefficient of Determination)**

---

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement3.csv')
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package in LPA')
```

```python
x = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
x_test
```

```python
y_test
```

```python
lr.predict(x_test.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(x_train,lr.predict(x_train),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
m = lr.coef_
```

```python
b = lr.intercept_
```
