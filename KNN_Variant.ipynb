# Multivariate Imputation 

- Multivariate imputation fills missing values using relationships between multiple features.

- It predicts missing values based on other variables in the dataset, not just a single column.

- More accurate when features are correlated.

# KNN Imputer 

- KNN Imputer is a multivariate imputation method based on K-Nearest Neighbors.

- For a missing value, it finds the K most similar data points (neighbors).

- Similarity is usually measured using distance metrics (e.g., Euclidean distance).

- The missing value is imputed using:

- Mean of neighbors (numerical data)

- Mode of neighbors (categorical data)

- Uses information from multiple features, not just the one with missing data.

# Advantages:

- Captures complex relationships between features

- Often more accurate than mean/median imputation

# Disadvantages:

- Computationally expensive for large datasets

- Sensitive to feature scaling

- Performance depends on choice of K

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('train.csv')[['Age','Pclass','Fare','Survived']]
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python

X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
knn = KNNImputer(n_neighbors = 3,weights = 'distance')

X_train_trf = knn.fit_transform(X_train)
X_test_trf = knn.transform(X_test)
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf,y_train)

y_pred = lr.predict(X_test_trf)

accuracy_score(y_test,y_pred)
```

```python
# Comparision with Simple Imputer --> mean

si = SimpleImputer()

X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf2,y_train)

y_pred2 = lr.predict(X_test_trf2)

accuracy_score(y_test,y_pred2)
```
